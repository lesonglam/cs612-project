```python
import pandas as pd
import os

# Starting from your original df
# df has columns: ['reviewerID', 'asin', 'reviewText', 'overall', 'feature', 'original_reviewText']

# 1. Drop rows that miss asin / overall / feature
df_clean = df.dropna(subset=['asin', 'overall', 'feature']).reset_index(drop=True)

# 2. Group by asin and compute:
#    - mean overall
#    - take the first feature (poster) for that movie
movies_df = (
    df_clean
    .groupby('asin', as_index=False)
    .agg({
        'overall': 'mean',   # mean rating
        'feature': 'first',  # pick one poster per movie
    })
)

# Optional: rename 'overall' to clarify that it's averaged
movies_df['overall'] = movies_df['overall'].astype(float)

print(movies_df.head())
print("Number of unique movies:", len(movies_df))

# 3. Export to a new file in a folder
os.makedirs('preprocessed', exist_ok=True)

# Option A  pickle (keeps Python objects like your nested 'feature' lists)
out_path_pkl = 'preprocessed/movies_mean_by_asin.pkl'
movies_df.to_pickle(out_path_pkl)

 
print("Saved:")
print("  ", out_path_pkl) 
```

```python
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad) / 1e6

from cnn import CNNConfig, CNNDataModule, CNNRegressor, seed_all

    # path_dataset = "shivamardeshna/movies-dataset/Data_s_E5"
path_dataset = "preprocessed/movies_mean_by_asin.pkl"

from kaggleloader import KaggleLoader
df = KaggleLoader(path_dataset).df()

# Configuration
model_name = "resnet18" 
cfg = CNNConfig(model_name=model_name, batch_size=256, epochs=50) 
model = CNNRegressor(cfg)
params_resnet = count_parameters(model)
print(f"ResNet18 parameters: {params_resnet:.2f} million")

model_name = "resnet50" 
cfg = CNNConfig(model_name=model_name, batch_size=256, epochs=50)
model = CNNRegressor(cfg)
params_resnet50 = count_parameters(model)
print(f"ResNet50 parameters: {params_resnet50:.2f} million")


model_name = "efficientnet_b0"    
cfg = CNNConfig(model_name=model_name, batch_size=256, epochs=50)
model = CNNRegressor(cfg)
params_effnet = count_parameters(model)
print(f"EfficientNet-B0 parameters: {params_effnet:.2f} million")
```
